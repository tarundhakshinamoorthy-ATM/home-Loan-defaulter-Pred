Feature Engineering

Number of Previous Loans

```python
bureau_features = bureau.groupby(
    "SK_ID_CURR"
).agg({

    "SK_ID_BUREAU":"count"

})

bureau_features.rename(

    columns={

        "SK_ID_BUREAU":"TOTAL_PREVIOUS_LOANS"

    },

    inplace=True

)
```

Active Loans

```python
active = bureau.groupby("SK_ID_CURR")["CREDIT_ACTIVE"]\
.apply(lambda x:(x=="Active").sum())

bureau_features["ACTIVE_LOANS"] = active
```

Closed Loans

```python
closed = bureau.groupby("SK_ID_CURR")["CREDIT_ACTIVE"]\
.apply(lambda x:(x=="Closed").sum())

bureau_features["CLOSED_LOANS"] = closed
```

Total Credit Amount

```python
bureau_features["TOTAL_CREDIT"] = bureau.groupby(
    "SK_ID_CURR"
)["AMT_CREDIT_SUM"].sum()
```

Average Credit

```python
bureau_features["AVG_CREDIT"] = bureau.groupby(
    "SK_ID_CURR"
)["AMT_CREDIT_SUM"].mean()
```

Maximum Credit

```python
bureau_features["MAX_CREDIT"] = bureau.groupby(
    "SK_ID_CURR"
)["AMT_CREDIT_SUM"].max()
```

Total Debt

```python
bureau_features["TOTAL_DEBT"] = bureau.groupby(
    "SK_ID_CURR"
)["AMT_CREDIT_SUM_DEBT"].sum()
```

Total Overdue

```python
bureau_features["TOTAL_OVERDUE"] = bureau.groupby(
    "SK_ID_CURR"
)["AMT_CREDIT_SUM_OVERDUE"].sum()
```

Average Overdue

```python
bureau_features["AVG_OVERDUE"] = bureau.groupby(
    "SK_ID_CURR"
)["AMT_CREDIT_SUM_OVERDUE"].mean()
```

Credit Utilization

```python
bureau_features["CREDIT_UTILIZATION"] = (

    bureau_features["TOTAL_DEBT"] /

    bureau_features["TOTAL_CREDIT"]

)
```

Debt Ratio

```python
bureau_features["DEBT_RATIO"] = (

    bureau_features["TOTAL_OVERDUE"] /

    bureau_features["TOTAL_DEBT"]

)
```

Days Since Last Credit

```python
bureau_features["LAST_CREDIT_DAYS"] = bureau.groupby(
    "SK_ID_CURR"
)["DAYS_CREDIT"].max()
```

Average Credit Duration

```python
bureau_features["AVG_CREDIT_DURATION"] = bureau.groupby(
    "SK_ID_CURR"
)["DAYS_CREDIT_ENDDATE"].mean()
```

Reset Index

```python
bureau_features.reset_index(inplace=True)
```

```python
df = df.merge(

    bureau_features,

    on="SK_ID_CURR",

    how="left"

)


```

Model Comparison

To identify the most suitable algorithm for predicting home loan defaults, four machine learning models were developed and evaluated: Logistic Regression, Balanced Logistic Regression, Random Forest, and XGBoost. Since the dataset was highly imbalanced, model evaluation was based not only on accuracy but also on precision, recall, F1-score, and ROC-AUC.

The first model developed was Logistic Regression, which served as the baseline for comparison. It achieved a high overall accuracy of nearly 92%, indicating that it correctly classified most applicants. However, a closer examination revealed that the model identified only a very small proportion of actual defaulters. This occurred because the model naturally favored the majority class (non-defaulters), making it less suitable for a real-world loan approval system where identifying risky customers is essential.

To overcome this limitation, Balanced Logistic Regression was implemented using class weights. Unlike the standard Logistic Regression model, this approach significantly improved the model's ability to identify defaulters. Although overall accuracy decreased, the model achieved the highest F1-score among all the evaluated models, indicating a much better balance between identifying risky customers and avoiding unnecessary false predictions. This makes Balanced Logistic Regression a strong candidate whenever detecting potential defaulters is the primary objective.

A Random Forest Classifier was then trained to capture complex nonlinear relationships within the data. Initially, the model heavily favored the majority class and predicted very few defaults. After introducing balanced class weights and adjusting the prediction threshold, the model became much more sensitive to default cases and achieved the highest recall among all the evaluated models. However, this improvement resulted in a larger number of false positives, reducing overall precision and accuracy. Although the model became effective at identifying risky customers, it tended to classify many safe applicants as potential defaulters.

Finally, XGBoost was trained as the advanced ensemble model. Among all the models evaluated, XGBoost consistently delivered the strongest overall predictive performance. Using the default prediction threshold, it achieved the highest accuracy, highest precision, and highest ROC-AUC score, demonstrating an excellent ability to distinguish between defaulters and non-defaulters. To further improve its practical usefulness, the prediction threshold was reduced from 0.50 to 0.30. This increased the number of correctly identified defaulters while maintaining strong overall classification performance. The threshold adjustment created a better balance between precision and recall without affecting the model's ranking capability.

Best Model for Production

Based on the overall evaluation, XGBoost with a decision threshold of 0.30 is recommended for production deployment. While Balanced Logistic Regression detected a larger proportion of defaulters, XGBoost provided the strongest overall performance by maintaining high accuracy, strong precision, the highest ROC-AUC score, and a noticeable improvement in recall after threshold optimization. Its ability to model complex financial relationships, combined with excellent predictive performance, makes it the most suitable choice for real-world home loan default prediction.

Conclusion

Each machine learning algorithm demonstrated unique strengths. Logistic Regression provided a reliable baseline, Balanced Logistic Regression significantly improved default detection, and Random Forest captured complex patterns while maximizing recall after threshold adjustment. However, XGBoost consistently outperformed the other models in terms of overall predictive capability. Considering both statistical performance and practical applicability, XGBoost with an optimized decision threshold is the recommended model for deployment in a production-level home loan risk assessment system.


# Challenges Faced and Techniques Used

## Introduction

Developing the Home Loan Default Prediction model involved much more than simply training machine learning algorithms. The project required integrating multiple datasets, handling missing information, engineering meaningful features, managing computational resources, and selecting an appropriate model for an imbalanced classification problem. Throughout the project, several practical challenges were encountered, each of which required a suitable data preprocessing or modeling technique to ensure reliable results.



## Handling Missing Values

One of the first challenges encountered was the presence of missing values across almost all datasets. Some columns contained only a few missing records, while others had more than half of their values missing. Using such incomplete information directly would have reduced model reliability.

To address this issue, columns with more than 50% missing values were removed because they contributed very little useful information. For the remaining numerical variables, missing values were replaced with the median, while categorical variables were filled using the most frequent category (mode). This approach helped preserve the overall distribution of the data while preventing unnecessary information loss.



## Working with Multiple Related Datasets

Unlike many machine learning projects that involve a single dataset, this project required combining information from seven different datasets. Each dataset represented a different aspect of the customer's financial history, including previous loans, bureau records, installment payments, POS transactions, and credit card behavior.

Since these datasets contained multiple records for each customer, directly merging them would have produced duplicate records and an extremely large dataset. Instead, meaningful customer-level features were created by aggregating information using statistical measures such as counts, averages, maximum values, totals, and ratios. These aggregated features were then merged into the main application dataset using the customer identifier.

This approach preserved valuable historical information while keeping the dataset manageable for model training.


## Class Imbalance

Another significant challenge was the highly imbalanced nature of the target variable. The majority of applicants were non-defaulters, while only a small percentage had actually defaulted on their loans.

Training a model directly on this data caused several algorithms to predict almost every applicant as a non-defaulter, resulting in high accuracy but poor identification of risky customers.

To overcome this issue, Balanced Logistic Regression and class-weighted Random Forest were implemented. In addition, the decision threshold of the XGBoost model was adjusted from 0.50 to 0.30, allowing the model to identify more potential defaulters while maintaining strong predictive performance. These techniques improved the practical usefulness of the final model.



## Feature Engineering

The original application data alone did not fully describe a customer's financial behaviour. Important indicators such as repayment history, overdue amounts, previous loan approvals, credit utilization, and payment delays were distributed across several supporting datasets.

To capture these behavioural patterns, extensive feature engineering was performed. New variables such as credit-to-income ratio, annuity-to-income ratio, payment ratios, credit utilization, late payment counts, repayment delays, and customer-level historical summaries were created. These engineered features allowed the machine learning models to learn relationships that were not immediately visible from the raw data.

Feature engineering proved to be one of the most important steps in improving the overall predictive performance of the models.



## Memory and Runtime Constraints

Working with large financial datasets created another practical challenge. Several datasets contained millions of records, and repeated merging operations frequently exhausted the available system memory, causing runtime interruptions.

To overcome this limitation, the project was divided into smaller stages. After completing feature engineering for each dataset, intermediate merged files were saved before moving to the next stage. Unnecessary variables were removed from memory using Python's `del` statement, followed by garbage collection (`gc.collect()`) to free RAM. This checkpoint-based workflow significantly reduced memory usage and prevented the need to repeat the entire preprocessing pipeline after unexpected runtime failures.



## Model Selection

No single machine learning model performed best across every evaluation metric. Some models achieved higher accuracy, while others detected a greater number of defaulters.

Instead of selecting the model based solely on accuracy, multiple evaluation metrics, including Precision, Recall, F1-score, and ROC-AUC, were considered. This provided a more comprehensive understanding of each model's strengths and weaknesses.

Although Balanced Logistic Regression achieved the highest F1-score and Random Forest detected the largest proportion of defaulters, XGBoost consistently demonstrated the strongest overall predictive capability. After optimizing the prediction threshold, XGBoost achieved an effective balance between identifying risky customers and maintaining high overall classification performance, making it the most appropriate model for production deployment.


## Conclusion

This project involved several real-world challenges commonly encountered in financial machine learning applications, including missing data, large-scale relational datasets, class imbalance, feature engineering, and computational limitations. By applying appropriate preprocessing techniques, careful feature engineering, and model optimization, these challenges were successfully addressed. The final solution demonstrates how combining data engineering with machine learning can produce a reliable and practical system for predicting home loan defaults.
